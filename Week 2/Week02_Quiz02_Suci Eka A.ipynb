```python
#Membuat scatterplot City dg SO2
import seaborn as sns
import matplotlib.pyplot as plt
data.plot.scatter(x = 'City', y = 'SO2', figsize=(10,5));
plt.xticks(rotation = 70)
plt.show()
```

**Plot Kota dengan SO2:**
    Berdasarkan hasil scatterplot antara kota dengan SO2 terlihat bahwa persebaran datanya tinggi, dimana masing-masing kota memiliki tingkat SO2 yang beragam. Hal itu tentunya mempengaruhi hasil klasterisasi karena tingginya persebaran data.
